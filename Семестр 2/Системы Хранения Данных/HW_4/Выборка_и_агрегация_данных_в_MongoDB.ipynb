### Упражнение 2. Конвейер агрегации

#### Задача 2.1 — **3 балла**

В рамках этой задачи требуется узнать, какие фильмы получили наибольшее число комментариев. Составьте для этого конвейер агрегации с перечисленными ниже этапами. Для удобства рекомендуется воспользоваться приложением Compass.

1. Сгруппировать документы коллекции `comments` по полю `movie_id` и подсчитать количество комментариев для каждого фильма, записав его в поле `count`.
1. Отсортировать получившийся набор данных по убыванию количества комментариев (так чтобы первый документ в этом наборе указывал на фильм с наибольшим количеством комментариев).
1. Используя оператор `$lookup`, присоединить коллекцию `movies` (по полю `movie_id`).
1. Выполнить проекцию, оставив в наборе данных только три поля: название фильма (`title`), год выпуска (`year`) и количество комментариев (поле `count`, добавленное на первом этапе). Для этого воспользуйтесь оператором `$project`: https://www.mongodb.com/docs/manual/reference/operator/aggregation/project/. Обратите внимание, что данные из коллекции `movies` (`title` и `year`) на этом этапе окажутся внутри массива.
1. Применить оператор `$unwind`, чтобы деконструировать этот массив. См. https://www.mongodb.com/docs/manual/reference/operator/aggregation/unwind/. В результате массив должен превратиться в одиночный объект.
1. Применить оператор `$addFields`, чтобы добавить поля `title` и `year` на верхний уровень структуры документа. См. https://www.mongodb.com/docs/manual/reference/operator/aggregation/addFields/.
1. Еще раз выполнить проекцию, чтобы удалить поле с массивом.

Ваш алгоритм решения может отличаться от приведенного здесь. Главное — достичь цели: получить плоский набор данных с тремя полями (`count` — количество комментариев, `year` — год выпуска, `title` — название фильма).

Скопируйте полученный конвейер в этот документ в виде кода на Python и выведите его на экран в виде списка словарей.

```python
# Ваш код здесь
comments_collection = db["comments"]

pipeline = [
    {
        "$group": {
            "_id": "$movie_id",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    },
    {
        "$lookup": {
            "from": "movies",
            "localField": "_id",
            "foreignField": "_id",
            "as": "movie_details"
        }
    },
    {
        "$project": {
            "movie_details.title": 1,
            "movie_details.year": 1,
            "count": 1,
            "_id": 0
        }
    },
    {
      "$unwind": "$movie_details"
    },
    {
        "$addFields": {
            "title": "$movie_details.title",
            "year": "$movie_details.year"
        }
    },
    {
        "$project": {
            "movie_details": 0
        }
    }
]
results = list(comments_collection.aggregate(pipeline))
'''
Выведем только первые десять фильмов
с наибольшим количеством количеством комментариев,
чтобы не замусоривать консоль
'''
for r in results[:10]:
  print(r)
```

#### Задача 2.2 — **1 балл**

Используя данные, полученные в задаче 2.1, выведите на экран столбчатую диаграмму, показывающую топ-20 наиболее комментируемых фильмов. По оси ординат выведите названия этих фильмов (подписи разместите слева диаграммы), по оси абсцисс — количество комментариев.

Для решения задачи можно использовать любую библиотеку, например Seaborn или Matplotlib.

```python
# Ваш код здесь
# График на чистом mathplotlib

import pandas as pd
import matplotlib.pyplot as plt

# Берем первые 20 записей из results
top_20 = results[:20]

# Сортировка по убыванию количества комментариев
top_20_sorted = sorted(top_20, key=lambda x: x['count'], reverse=True)

# Создаем списки для визуализации
titles = [movie['title'] for movie in top_20_sorted]
counts = [movie['count'] for movie in top_20_sorted]

# Построение диаграммы
plt.figure(figsize=(10, 8))
plt.barh(titles, counts, color='dodgerblue')  # Используем более контрастный цвет
plt.xlabel('Количество комментариев', fontsize=12)
plt.ylabel('Название фильма', fontsize=12)
plt.title('Топ-20 наиболее комментируемых фильмов', fontsize=16, fontweight='bold')

# Инвертируем ось Y, чтобы фильмы с наибольшим количеством комментариев были вверху
plt.gca().invert_yaxis()

# Добавляем сетку для улучшения читаемости
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Улучшаем отображение
plt.tight_layout()
plt.show()

```

```python
# График с использованием seaborn
'''
На предыдущем графике заметно, что разница между минимальным и максимальным
количеством комментариев в выборке минимальное и составляет всего 8 при общем
количестве комментариев более 150.
Для того, чтобы разница была более очевидной реализуем масштабированную шкалу.
'''

import seaborn as sns
import numpy as np


# Создаем DataFrame для визуализации и сортируем его
df = pd.DataFrame(results[:20])
df = df.sort_values(by='count', ascending=True)

min_count = df["count"].min()
max_count = df["count"].max()

# Используем цветовую схему Dark2_r, так как у него нет градиента,
# и белые цифры на нем выглядят лучше
colors = plt.cm.Dark2_r(np.linspace(0, 1, len(df)))

plt.figure(figsize=(14, 10))
ax = plt.gca()

# Устанавливаем минимальное значение оси X чуть меньше минимального значения данных
min_x = df["count"].min() * 0.98

# Создаем горизонтальную гистограмму с минимальным значением, отличным от нуля
bars = plt.barh(df["title"], df["count"], color=colors)
plt.xlim(min_x, df["count"].max()+1)

x_ticks = np.linspace(min_x, max_count, 10)
plt.xticks(x_ticks, [f"{int(x)}" for x in x_ticks], fontsize=10)

# Добавление текстовых меток на столбцы
for bar in bars:
    width = bar.get_width()
    plt.text(
        width,
        bar.get_y() + bar.get_height() / 2,
        f'{int(width)}',
        va='center',
        ha='right',
        color='white',
        fontsize=10,
        fontweight='bold'
    )

# Настройка заголовка и осей
plt.title("Топ-20 наиболее комментируемых фильмов\n(Масштабированная шкала)",
          fontsize=16,
          fontweight='bold')
plt.xlabel("Количество комментариев", fontsize=12)
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Рендеринг графика
plt.show()
```
